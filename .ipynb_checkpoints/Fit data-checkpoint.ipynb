```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
theta = np.linspace(0,pi/2,11)
alpha3 = 10000000
gamma3 = 500
w_B = 0.05
B_0 = 1
omega = 2*pi/66.60e-6
T = 4*pi/(0.05* omega/np.sin(theta))
```

```python
kapa_se = np.cos(theta)**2 * alpha3 * gamma3 * T**3 / 12 + w_B**2 * np.sin(theta)**4 * alpha3 * T**2 / (2*B_0**2)
kapa_scheme1_se = (np.cos(theta)-w_B*np.sin(theta)**2/B_0)**2 * alpha3*gamma3*T**3/12
kapa_scheme1_cpmg = (np.cos(theta)-w_B*np.sin(theta)**2/B_0)**2 * alpha3*gamma3*T**3/48
kapa_scheme2_se = np.cos(theta)**2 * alpha3 * gamma3 * T**3 / 48 + w_B**2 * np.sin(theta)**4 * alpha3 * gamma3 * T**3 / (12*B_0**2)
```

```python
w_se = 1.0 / np.exp(kapa_se)
w_scheme1_se = 1.0 / np.exp(kapa_scheme1_se)
w_scheme1_cpmg = 1.0 / np.exp(kapa_scheme1_cpmg)
w_scheme2_se = 1.0 / np.exp(kapa_scheme2_se)
```

```python
plt.plot(theta,w_se,'b',theta,w_scheme1_se,'r',theta,w_scheme1_cpmg,'g',theta,w_scheme2_se,'b')
```
