# Set Berry Phase Wave Function

## Single Run

```python
def ExperimentWave(theta, T, RabiTime, omega_rf, delta, Amp=1.0):

    frequency = 2 * pi * 200e6
    clock_amp = 1.0
    U_time = theta / (2 * pi) * RabiTime
    U_time = round(U_time, 7)

    t0 = 0
    t1 = t0 + RabiTime / 4
    t2 = t1 + U_time
    t3 = t2 + T
    t4 = t3 + RabiTime / 2
    t5 = t4 + 2 * U_time
    t6 = t5 + T
    t7 = t6 + RabiTime / 2
    t8 = t7 + 2 * U_time
    t9 = t8 + U_time
    t10 = t9 + RabiTime
    t_total = t10

    N = int(t_total / dt) + 1
    t = np.arange(0, N) * dt

    wave_x = np.zeros(N)
    wave_y = np.zeros(N)

    # set t0 to t1, prepare superposition |0>+|1>
    idx = np.argwhere((t >= t0) & (t < t1))
    wave_x[idx] = clock_amp * np.cos(frequency * t[idx] + pi / 2)
    wave_y[idx] = clock_amp * np.cos(frequency * t[idx] + pi / 2)
    phase1 = frequency * t1

    # set t1 to t2 , U.dag operation to eigenstate of H
    idx = np.argwhere((t >= t1) & (t < t2))
    wave_x[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase1 + pi / 2)
    wave_y[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase1 + pi / 2)
    phase2 = phase1 + frequency * (t2 - t1)

    # set t2 to t3 , adiabatic operation 1
    idx = np.argwhere((t >= t2) & (t < t3))
    phi_1 = -omega_rf * (t[idx] - t[idx][0])
    wave_x[idx] = clock_amp * np.cos(
        (frequency + delta) * (t[idx] - t[idx][0]) + phase2 + pi + phi_1)
    wave_y[idx] = clock_amp * np.cos(
        (frequency + delta) * (t[idx] - t[idx][0]) + phase2 + pi + phi_1)
    phase3 = phase2 + (frequency + delta) * (t3 - t2)

    # set t3 to t4 , exchange eigenstate,step 1
    idx = np.argwhere((t >= t3) & (t < t4))
    wave_x[idx] = clock_amp * np.cos(frequency * (t[idx] - t[idx][0]) + phase3)
    wave_y[idx] = clock_amp * np.cos(frequency * (t[idx] - t[idx][0]) + phase3)
    phase4 = phase3 + frequency * (t4 - t3)

    # set t4 to t5 , exchange eigenstate,step 2
    idx = np.argwhere((t >= t4) & (t < t5))
    wave_x[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase4 + pi / 2)
    wave_y[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase4 + pi / 2)
    phase5 = phase4 + frequency * (t5 - t4)

    # set t5 to t6 , adiabatic operation 2
    idx = np.argwhere((t >= t5) & (t < t6))
    phi_2 = omega_rf * (t[idx] - t[idx][0])
    wave_x[idx] = clock_amp * np.cos(
        (frequency + delta) * (t[idx] - t[idx][0]) + phase5 + pi + phi_2)
    wave_y[idx] = clock_amp * np.cos(
        (frequency + delta) * (t[idx] - t[idx][0]) + phase5 + pi + phi_2)
    phase6 = phase5 + (frequency + delta) * (t6 - t5)

    # set t6 to t7 , exchange eigenstate,step 1
    idx = np.argwhere((t >= t6) & (t < t7))
    wave_x[idx] = clock_amp * np.cos(frequency * (t[idx] - t[idx][0]) + phase6)
    wave_y[idx] = clock_amp * np.cos(frequency * (t[idx] - t[idx][0]) + phase6)
    phase7 = phase6 + frequency * (t7 - t6)

    # set t7 to t8 , exchange eigenstate,step 2
    idx = np.argwhere((t >= t7) & (t < t8))
    wave_x[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase7 + pi / 2)
    wave_y[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase7 + pi / 2)
    phase8 = phase7 + frequency * (t8 - t7)

    # set t8 to t9 , U operation to eigenstate of H
    idx = np.argwhere((t >= t8) & (t < t9))
    wave_x[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase8 - pi / 2)
    wave_y[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase8 - pi / 2)
    phase9 = phase8 + frequency * (t9 - t8)

    # set t9 to t10, tomography rotation
    idx = np.argwhere((t >= t9) & (t < t10))
    wave_x[idx] = clock_amp * np.cos(frequency *
                                     (t[idx] - t[idx][0]) + phase9 + pi / 2)
    wave_y[idx] = clock_amp * np.cos(frequency * (t[idx] - t[idx][0]) + phase9)

    return (wave_x, wave_y), t_total - RabiTime
```

```python
for i in range(1,21):
    dt = 1e-9
    RabiTime = 24.28e-05
    Omega_0 = 2*pi/ RabiTime
    omega_rf = 2*pi*5e2  # 1k
    theta = i*pi/40
    B0 = Omega_0/sin(theta)
    delta = Omega_0/sin(theta)   # 10k
    delta = round(delta,0)
    # B0 = sqrt(delta**2 + Omega_0**2)
    # theta = np.arccos(delta/B0) 
    delta_t = 2*pi/omega_rf/100
    T = 2*pi/omega_rf
    T = round(T,7)
    print 'cos(theta):', np.cos(theta)
    print 'theta:',theta
    print 'Berry Phase:', -4*pi*np.cos(theta)
    print 'B0:',B0

    # 微波基本参数
    amp1 = 0.1
    amp2 = 0.1   
    dt =1.0e-9                 # AFG sampling Rate
    omega_1 = 200e6 * 2 * pi # clock transition  frequency

    ####################################################################################  生成波形  
    # SequTime 已经包含Tomegraphy的波形，此处通过波形的时间来确定操作的时间，'Sz0'等都是>微波时间的
    Wave,SequTime = ExperimentWave(theta, T, RabiTime, omega_rf, delta, Amp=1.0)
    wave_x,wave_y = Wave
    ######################################################################################

    SequTime = SequTime * 1e6
    PiTime = RabiTime/2.0 * 1e6
    times = {'Sz0': SequTime,                   # P0的概率，下同
             'Sz1': SequTime + PiTime,
             'Sx0': SequTime + PiTime/2.0,
             'Sy0': SequTime + PiTime/2.0,
             'Sx1': SequTime + 3*PiTime/2.0,
             'Sy1': SequTime + 3*PiTime/2.0,
             }

    code = strHeader + strSeq.format(**times) + strTask + strPlots + strDataFolder

    ion.setCode(code)

    time.sleep(1)

    wave1 = np.vstack((wave_x,wave_y))
    wave2 = wave1   
    print(ion.cmd('TaskFun', 'setWave',((wave1,wave2),(amp1,amp2)))) # 两个波形轮流执行
    title = code
    fname = 'experment'+'_theta='+str(round(theta,2))
    recorder = PickleRecorder(title, port=data_port,
                                     fpath = './data/zeem_energy',
                                     fname = fname)
    recorder.run()
    print(ion.cmd('start'))

    while True:
        data_num = recorder.i
        print(data_num)
        if data_num > 9: # 9的意思是执行10轮
            break
        time.sleep(1)
    recorder.stop()
    #         time.sleep(10)
    ion.cmd('stop')
```

# Zeeman energy

```python
def ExperimentWave_scheme1(theta_a, theta_c, T, RabiTime_a, RabiTime_c, omega_rf, delta_a, delta_c, Amp_a=0.3, Amp_c=0.55):

    frequency = 2 * pi * 200e6
    Ua_time = theta_a/(2*pi)*RabiTime_a
    Ua_time = round(Ua_time,7)
    Uc_time = theta_c/(2*pi)*RabiTime_c
    Uc_time = round(Uc_time,7)
    
    t0 = 0
    t1 = t0 + RabiTime_a/4
    t2 = t1 + Ua_time
    t3 = t2 + T
    t4 = t3 + RabiTime_a/2
    t5 = t4 + 2*Ua_time
    t6 = t5 + T
    t7 = t6 + RabiTime_c/2
    t8 = t7 + 2*Uc_time
    t9 = t8 + Uc_time
    t10 = t9 + RabiTime_c
    t_total = t10

    N = int(t_total/dt)+1
    t = np.arange(0,N)*dt
    
    wave_x = np.zeros(N)
    wave_y = np.zeros(N)

    # set t0 to t1, prepare superposition |0>+|1>
    idx = np.argwhere((t >= t0) & (t < t1))
    wave_x[idx] = Amp_a * np.cos(frequency * t[idx] + pi/2)
    wave_y[idx] = Amp_a * np.cos(frequency * t[idx] + pi/2)
    phase1 = frequency * t1

    # set t1 to t2 , U.dag operation to eigenstate of H
    idx = np.argwhere((t >= t1) & (t < t2))
    wave_x[idx] = Amp_a * np.cos(frequency * (t[idx] - t[idx][0]) + phase1 + pi/2)
    wave_y[idx] = Amp_a * np.cos(frequency * (t[idx] - t[idx][0]) + phase1 + pi/2)
    phase2 = phase1 + frequency*(t2-t1)
    
    # set t2 to t3 , adiabatic operation 1
    idx = np.argwhere((t >= t2) & (t < t3))
    phi_1 = -omega_rf*(t[idx] - t[idx][0])
    wave_x[idx] = Amp_a * np.cos((frequency + delta_a) * (t[idx] - t[idx][0]) + phase2 + pi + phi_1)
    wave_y[idx] = Amp_a * np.cos((frequency + delta_a) * (t[idx] - t[idx][0]) + phase2 + pi + phi_1)
    phase3 = phase2 + (frequency + delta_a) *(t3-t2)
    
    # set t3 to t4 , exchange eigenstate,step 1
    idx = np.argwhere((t >= t3) & (t < t4))
    wave_x[idx] = Amp_a * np.cos(frequency * (t[idx] - t[idx][0]) + phase3)
    wave_y[idx] = Amp_a * np.cos(frequency * (t[idx] - t[idx][0]) + phase3)
    phase4 = phase3 + frequency *(t4-t3)
    
    # set t4 to t5 , exchange eigenstate,step 2
    idx = np.argwhere((t >= t4) & (t < t5))
    wave_x[idx] = Amp_a * np.cos(frequency * (t[idx] - t[idx][0]) + phase4 + pi/2)
    wave_y[idx] = Amp_a * np.cos(frequency * (t[idx] - t[idx][0]) + phase4 + pi/2)
    phase5 = phase4 + frequency*(t5-t4)
    
    # set t5 to t6 , adiabatic operation 2
    idx = np.argwhere((t >= t5) & (t < t6))
    phi_2 = omega_rf*(t[idx] - t[idx][0])
    wave_x[idx] = Amp_c * np.cos((frequency + delta_c) * (t[idx] - t[idx][0]) + phase5 + pi + phi_2)
    wave_y[idx] = Amp_c * np.cos((frequency + delta_c) * (t[idx] - t[idx][0]) + phase5 + pi + phi_2)
    phase6 = phase5 + (frequency + delta_c) *(t6-t5)
    
    # set t6 to t7 , exchange eigenstate,step 1
    idx = np.argwhere((t >= t6) & (t < t7))
    wave_x[idx] = Amp_c * np.cos(frequency * (t[idx] - t[idx][0]) + phase6)
    wave_y[idx] = Amp_c * np.cos(frequency * (t[idx] - t[idx][0]) + phase6)
    phase7 = phase6 + frequency *(t7-t6)
    
    # set t7 to t8 , exchange eigenstate,step 2
    idx = np.argwhere((t >= t7) & (t < t8))
    wave_x[idx] = Amp_c * np.cos(frequency * (t[idx] - t[idx][0]) + phase7 + pi/2)
    wave_y[idx] = Amp_c * np.cos(frequency * (t[idx] - t[idx][0]) + phase7 + pi/2)
    phase8 = phase7 + frequency*(t8-t7)
    
    # set t8 to t9 , U operation to eigenstate of H
    idx = np.argwhere((t >= t8) & (t < t9))
    wave_x[idx] = Amp_c * np.cos(frequency * (t[idx] - t[idx][0]) + phase8 - pi/2)
    wave_y[idx] = Amp_c * np.cos(frequency * (t[idx] - t[idx][0]) + phase8 - pi/2)
    phase9 = phase8 + frequency*(t9-t8)

    # set t9 to t10, tomography rotation
    idx = np.argwhere((t >= t9) & (t < t10))
    wave_x[idx] = Amp_c * np.cos(frequency * (t[idx] - t[idx][0]) + phase9 + pi / 2)
    wave_y[idx] = Amp_c * np.cos(frequency * (t[idx] - t[idx][0]) + phase9) 

    return (wave_x,wave_y), t_total-RabiTime_c
```

```python
# parameters
dt = 1e-9
RabiTime_a = 30.00e-05
Omega_0 = 2*pi/ RabiTime_a
omega_rf = 2*pi*3e2  # 1k
delta_a = 2*pi*3e3   # 10k
B0 = sqrt(delta_a**2 + Omega_0**2)
theta_a = np.arccos(delta_a/B0) 
theta_c = np.arccos((cos(theta_a)-2*omega_rf/B0+omega_rf**2*cos(theta_a)/(B0**2+omega_rf**2))\
           /(1+(omega_rf**2-2*omega_rf*B0*cos(theta_a))/(B0**2+omega_rf**2)))
RabiTime_c = 2*pi/(B0*sin(theta_c))
RabiTime_c = round(RabiTime_c,7)
delta_c = B0*cos(theta_c)
T = 2*pi/omega_rf
T = round(T,7)
print 'cos(theta_a):', np.cos(theta_a)
print 'theta_a:',theta_a
print 'Berry Phase:', -2*pi*np.cos(theta_a) - 2*pi*np.cos(theta_c)
print 'RabiTime_c:', RabiTime_c
```
